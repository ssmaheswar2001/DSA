```python
# Time Complexity : O(ElogE)

import heapq
def spanningTree(V, edges):

    adj = [[] for i in range(V)]

    for i in edges:
        u, v, weight = i
        adj[u].append((v, weight))
        adj[v].append((u, weight))
    print(adj)
    
    visited = [0 for i in range(V)]
        
    sum = 0
    MST = []
    
    pQueue = []
    heapq.heappush(pQueue, (0, 0, -1))
    
    while len(pQueue) > 0:
        distance, u, parent = heapq.heappop(pQueue)
        if visited[u] == 0:
            visited[u] = 1
            if parent != -1:
                MST.append((parent, u, distance))
                sum += distance
        for i in adj[u]:
            v, weight = i
            if visited[v] == 0 :
                heapq.heappush(pQueue, (weight, v, u))
    print(MST) 
    return sum

V = 5
edges = [[0, 1, 2], [0, 2, 1], [1, 2, 1], [2, 3, 2], [3, 4, 1], [4, 2, 2]]
spanningTree(V, edges)
```
