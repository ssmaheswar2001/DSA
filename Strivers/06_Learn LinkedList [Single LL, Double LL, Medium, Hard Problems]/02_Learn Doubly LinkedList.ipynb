# Array to DoubleLinkedList

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
class DoubleLL:
    def __init__(self):
        self.head = None
    
    def constructDLL(self, arr):
        n = len(arr)
        
        self.head = Node(arr[0])
        current = self.head
        
        for i in range(1, n):
            newNode = Node(arr[i])
            current.next = newNode
            newNode.prev = current
            current = current.next
            
        return self.head
    
    def printDLL(self, head):
        current = head
        
        while current != None:
            print(current.data, end='<->')
            current = current.next
            
        print()
    
arr = [1, 2, 3, 4, 5]
dll = DoubleLL()
head = dll.constructDLL(arr)
dll.printDLL(head)
```

# Deletion of Node in DoubleLinked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
class DoubleLL:
    def __init__(self):
        self.head = None
    
    def constructDLL(self, arr):
        n = len(arr)
        
        self.head = Node(arr[0])
        current = self.head
        
        for i in range(1, n):
            newNode = Node(arr[i])
            current.next = newNode
            newNode.prev = current
            current = current.next
            
        return self.head
    
    def deleteHead(self, head):
        if head is None:
            return None
        if head.next is None:
            return None
        
        current = head
        head = current.next
        head.prev = None
        current.next = None
        return head
    
    def deleteTail(self, head):
        if head is None or head.next is None:
            return None
        
        current = head
        while current.next != None:
            current = current.next
        
        previousNode = current.prev
        current.prev = None
        previousNode.next = None
        return head
    
    def deleteKElement(self, head, k):
        if head == None:
            return None
        
        current = head
        count = 0
        while current != None:
            count += 1
            if count == k:
                break
            current = current.next
            
        if current.prev == None and current.next == None:
            return None
        if current.prev == None:
            return self.deleteHead(head)
        if current.next == None:
            return self.deleteTail(head)
        
        previousNode = current.prev
        nextNode = current.next
        
        previousNode.next = nextNode
        nextNode.prev = previousNode
        
        current.prev = None
        current.next = None
        return head
        
    
    def printDLL(self, head):
        current = head
        while current != None:
            print(current.data, end='<->')
            current = current.next
        print()
    
arr = [1, 3, 4]
dll = DoubleLL()
head = dll.constructDLL(arr)
dll.printDLL(head)
head = dll.deleteKElement(head, 3)
dll.printDLL(head)
```

# Reverse A DLL

```python
# Better Approach : Using Stack DS
# Time Complexity : O(N) + O(N)
# Space Complexity : O(N)

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
class DoubleLL:
    def __init__(self):
        self.head = None
    
    def constructDLL(self, arr):
        n = len(arr)
        
        self.head = Node(arr[0])
        current = self.head
        
        for i in range(1, n):
            newNode = Node(arr[i])
            current.next = newNode
            newNode.prev = current
            current = current.next
            
        return self.head
    
    def reverseDLL(self, head):
        
        stack = []
        current = head
        
        while current != None:
            stack.append(current.data)
            current = current.next
        
        top = len(stack)-1
        current = head
        while current != None:
            current.data = stack[top]
            top -= 1
            current = current.next
        return head
    
    def printDLL(self, head):
        current = head
        
        while current != None:
            print(current.data, end='<->')
            current = current.next
            
        print()
    
arr = [1, 2, 3, 4, 5]
dll = DoubleLL()
head = dll.constructDLL(arr)
dll.printDLL(head)
head = dll.reverseDLL(head)
dll.printDLL(head)
```
